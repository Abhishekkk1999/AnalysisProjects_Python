```python
import numpy as np
import pandas as pd
import plotly.plotly as pl
import plotly.offline as po
import cufflinks as cf
```

```python
po.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
def createdata(data):
    if(data == 1):
        x = np.random.rand(100,5)
        df1 = pd.DataFrame(x, columns=['A','B','C','D','E'])
    elif(data == 2):
        x = [0,0,0,0,0]
        r1 = [0,0,0,0,0]
        r2 = [0,0,0,0,0]
        r3 = [0,0,0,0,0]
        r4 = [0,0,0,0,0]
        print("Enter the value of Columns")
        i = 0
        for i in [0,1,2,3,4]:
            x[i] = input()
            i = i + 1
        print("Enter the value for First Row")
        i = 0
        for i in [0,1,2,3,4]:
            r1[i] = input()
            i = i + 1
        print("Enter the value for Second Row")
        i = 0
        for i in [0,1,2,3,4]:
            r2[i] = input()
            i = i + 1
        print("Enter the value for Third Row")
        i = 0
        for i in [0,1,2,3,4]:
            r3[i] = input()
            i = i + 1
        print("Enter the value for Fourth Row")
        i = 0
        for i in [0,1,2,3,4]:
            r4[i] = input()
            i = i + 1
        
        df1 = pd.DataFrame([r1,r2,r3,r4], columns = x)
    elif(data == 3):
        file = input("Input the file name")
        x = pd.read_csv(file)
        df1 = pd.DataFrame(x)
    else:
        print("DataFrame creation failed please enter something between 1 to 3 and try again")
    return df1
        
             
```

```python
def plotter(plot):
    if(plot == 1):
        finalplot = df1.iplot(kind = "scatter") 
    elif(plot == 2):
        finalplot = df1.iplot(kind = "scatter", mode = 'markers', symbol = 'x', colorscale = 'paired')
    elif(plot == 3):
        finalplot = df1.iplot(kind = "bar")
    elif(plot == 4):
        finalplot = df1.iplot(kind = 'hist')
    elif(plot == 5):
        finalplot = df1.iplot(kind = 'box')
    elif(plot == 6):
        finalplot = df1.iplot(kind = 'surface')
    else:
        finalplot = print("Select only between 1 to 7 ❌")
    return finalplot      
```

```python
def plotter2(plot):
    col = input("Enter the no. of columns you want to plot by selecting only 1,2 or 3")
    col = int(col)
    if(col == 1):
        colm = input("Enter the column you want to plot by selecting any column from dataframe head")
        if(plot == 1):
            finalplot = df1[colm].iplot(kind =  'scatter')
        elif(plot == 2):
            finalplot = df1[colm].iplot(kind =  'scatter' , mode = 'markers', symbol = 'x', colorscale = 'paired')
        elif(plot == 3):
            finalplot = df1[colm].iplot(kind = 'bar')
        elif(plot == 4):
            finalplot = df1[colm].iplot(kind = 'hist')
        elif(plot == 5):
            finalplot = df1[colm].iplot(kind = 'box')
        elif(plot == 6 or plot == 7):
            finalplot = print("Bubble plot or Surface Plot require more than one column arguments ☹")
        else:
            finalplot = print("Select only between 1 to 7 ❌") 
    
    elif(col == 2):
        print("Enter the columns you want to plot by selecting from dataframe head")
        x = input("First Column")
        y = input("Second Column")
        if(plot == 1):
            finalplot = df1[[x,y]].iplot(kind =  'scatter')           
        elif(plot == 2):
            finalplot = df1[[x,y]].iplot(kind =  'scatter' , mode = 'markers', symbol = 'x', colorscale = 'paired')
        elif(plot == 3):
            finalplot = df1[[x,y]].iplot(kind =  'bar') 
        elif(plot == 4):
            finalplot = df1[[x,y]].iplot(kind =  'hist') 
        elif(plot == 5):
            finalplot = df1[[x,y]].iplot(kind =  'box') 
        elif(plot == 6):
            finalplot = df1[[x,y]].iplot(kind =  'surface') 
        elif(plot == 7):
            size = input("Please enter the size of the column for bubble plot")
            finalplot = df1.iplot(kind =  'bubble', x=x, y=y, size = size) 
        else:
            print("Select only between 1 to 7 ❌") 
            
    elif(col == 3):
        print("Enter the columns you want to print")
        x = input("First Column")
        y = input("Second Column")
        z = input("Third Column")
        if(plot == 1):
            finalplot = df1[[x,y,z]].iplot(kind =  'scatter')           
        elif(plot == 2):
            finalplot = df1[[x,y,z]].iplot(kind =  'scatter' , mode = 'markers', symbol = 'x', colorscale = 'paired')
        elif(plot == 3):
            finalplot = df1[[x,y,z]].iplot(kind =  'bar') 
        elif(plot == 4):
            finalplot = df1[[x,y,z]].iplot(kind =  'hist') 
        elif(plot == 5):
            finalplot = df1[[x,y,z]].iplot(kind =  'box') 
        elif(plot == 6):
            finalplot = df1[[x,y,z]].iplot(kind =  'surface') 
        elif(plot == 7):
            size = input("Please enter the size of the column for bubble plot")
            finalplot = df1.iplot(kind =  'bubble', x=x, y=y,z=z, size = size) 
        else:
            print("Select only between 1 to 7 ❌") 
            
    else:
        finalplot = print("Please enter 1,2 or 3")
    return finalplot
    
    
```

```python
def main(cat):
    if(cat == 1):
        print("Select the type of plot you need to plot by writing 1 to 6")
        print("1. Line Plot")
        print("2. Scatter Plot")
        print("3. Bar Plot")
        print("4. Histogram Plot")
        print("5. Box Plot")
        print("6. Surface Plot")
        plot = int(input())
        output = plotter(plot)
    elif(cat == 2):
        print("Select the type of plot you need to plot by writing 1 to 7")
        print("1. Line Plot")
        print("2. Scatter Plot")
        print("3. Bar Plot")
        print("4. Histogram Plot")
        print("5. Box Plot")
        print("6. Surface Plot")
        print("7. Bubble Plot")
        plot = int(input())
        output = plotter2(plot) 
    else:
        print("Please enter 1 or 2 and try again ☠")
        
```
